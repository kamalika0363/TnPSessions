```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
# root -> 
class LinkedList:
    def createLinkedList(self):
        self.root = None # Root is not created but assigned
    def insertLeft(self, e):
        n = Node(e)
        if self.root==None:
            self.root = n # if not there assign n as root        
        else:
            n.next=self.root #1
            self.root=n #2
        print(e, "inserted at left")      

    def deleteLeft(self):
        if self.root==None:
            print("Empty List")    
        else:
            t=self.root #1
            self.root=self.root.next #2
        print(t.data, "deleted from list")        
        
    def insertRight(self, e):
        n = Node(e)
        if self.root==None:
            self.root = n # if not there assign n as root        
        else:
            t=self.root #1
            while t.next!=None:
                t=t.next #2
            t.next = n #3
        print(e, "inserted at right")        
        
    def deleteRight(self):
        if self.root==None:
            print("list is empty")
        else:
            t2=t=self.root #1
            while t.next!=None: #2
                t2=t
                t=t.next 
            if self.root == t:
                self.root=None
            else:
                t2.next=None #3                    
        print(t.data, "deleted from list")       
        
    def printList(self):
        if self.root==None:
            print("list is empty")
        else:
            t=self.root
            while t!=None:
                print(t.data, end="->")
                t=t.next
            print()     
            
    def searchList(self,key):
        if self.root==None:
            print("list is empty")
        else:
            # WHILE ELSE
            t=self.root
            while t!=None: #2
                if key == t.data:
                    print(key,"Found...")
                    break
                t=t.next
            else: # only executes if while runs completely
                print(key,"Not Found...")
                 
            
            
obj=LinkedList()
obj.createLinkedList()
while True:
    ch=int(input("\n1.Insert Left\n2.Delete Left\n3.Insert Right\n4.Delete Right\n5.Display\n0.Exit:"))
    if ch==1:
        e=int(input("Enter data:"))
        obj.insertLeft(e)
    elif ch==2:
        obj.deleteLeft()
    elif ch==3:
        e=int(input("Enter data:"))
        obj.insertRight(e)
    elif ch==4:
        obj.deleteRight()
    elif ch==5:
        obj.printList()
    elif ch==6:
        obj.searchList(int(input("Enter key to search:")))        
    elif ch==0:
         print("THANKS FOR USING THE CODE")
         break
    else:
        print("Wrong option given")                             
```

# Summary

- `createLinkedList(self)`: This method initializes the linked list with a `root` node set to `None`. This means the list is empty at the start.

- `insertLeft(self, e)`: This method inserts a new node to the left (beginning) of the linked list. If the list is empty, it sets the new node as the root. Otherwise, it sets the new node's `next` pointer to the current root and then updates the root to be the new node.

- `deleteLeft(self)`: This method deletes a node from the left (beginning) of the linked list. If the list is empty, it prints "Empty List". Otherwise, it moves the root pointer to the next node, effectively removing the first node from the list.

- `insertRight(self, e)`: This method inserts a new node to the right (end) of the linked list. If the list is empty, it sets the new node as the root. Otherwise, it traverses to the end of the list and sets the last node's `next` pointer to be the new node.

- `deleteRight(self)`: This method deletes a node from the right (end) of the linked list. If the list is empty, it prints "list is empty". Otherwise, it traverses to the end of the list and removes the last node by setting the second last node's `next` pointer to `None`.

- `printList(self)`: This method prints all elements in the linked list from left (beginning) to right (end). If the list is empty, it prints "list is empty". Otherwise, it traverses through each node and prints its data.
